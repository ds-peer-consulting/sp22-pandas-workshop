Being able to manipulate those dataframes is crucial for data scientists to be able to clean up their data in a way that is usable for analysis.

````python
# the conventional Pandas import statement
import pandas as pd
import numpy as np
````

##### What is a DataFrame and how does Series work?

A DataFrame is like a 2D table of information. Each row is called a record or observation because rows typically represent some real life event or object. And each column is called a feature because it's an aspect of the real life event or object.
A Series is a special data type that represents the column of a dataframe. 

A Series works very similar to a NumPy because it is literally a NumPy array with some extra functions built into it. More information about Series is always a Google search away, if you are interested in learning more about those extra functions.


There are a few different ways to add a columns to a dataframe using Pandas, but we can start by looking at an example of a small table.

````python
df = pd.DataFrame({"A": [1, 2, 3, 4],
                   "B": [5, 6, 7, 8]})
df
````

Now lets say we wanted to search for something specific in the data frame.We can use both **df.loc** and **df.iloc**.
##### loc vs iloc?
df.loc is a label based search and df.iloc is an integer position based search.
Both utilize the followign syntax:
~~~~
loc[row_label, column_label]
iloc[row_position, column_position]
~~~~

With labels you can locate an item via the column and row name (note: for the below example it would typically be df.loc['row name', 'column name'] where both are surrounded by an apostrophe, however given there is no row name, an integer will suffice)"

````python
df.loc[1,'A']
````

````python
df.loc[1]
````

With integers you can locate the ith row and ith position as such

````python
df.iloc[0,1]
````

Or with a list of integers

````python
df.iloc[[0, 1], [1, 1]]
````

You can even use slices

````python
df.iloc[:2]
````

The first way we can look at it is simply using the format df['New_Column_Name'] = [Array]

````python
df["New_Column"] = [1, 2, 3, 4]
df
````

We also can use the built in function $.insert()$ to specify the index we want to place the column. **Reminder that dataframes are zero indexed!**

````python
df.insert(2, "C", [21, 23, 24, 21])
df
````

*(Bonus)* Now here is an example of how can combine a previous method of adding columns that we discussed with a dictionary to more quickly fill in the data.

````python
dictionary = {'A': '1', 'B': '2',
           'C': '3', 'D': '4'}

df['alphabet'] = dictionary
df
````

We also can look out how we can rename columns once the information we want is added to the dataframe.

````python
df2 = df.rename(columns={'A': 'X', 'B': 'Y'})  
df2
````

````python
#There is the option to specify the index‚Äìwhich can be useful when dealing with multi-index tables (i.e., pivot tables)
df2.rename({'A':'x1', 'B':'x2', 'C':'x3', 'New_Column':'x4', 'alphabet':'x5'}, axis=1)
````

If you look you can see the first two columns do not change when we try to rename them. This is because the variable is stored and we will need to clear variables if we would like to change it again.

````python
#You can also assign column names directly
df.columns = ['V', 'W', 'X', 'Y', 'Z']
df
````

Next we will look at how to combine various data frames into a single dataframe.

````python
new_df1 = pd.DataFrame({"A": ['A', 'B', 'C', 'D'],
                   "B": [10, 20, 30, 40]})
new_df2 = pd.DataFrame({"A": ['A', 'B', 'C', 'D'],
                   "B": [50, 60, 70, 80]})
new_df1, new_df2
````

````python
#The first method of concat will add all rows from the dataframes together, placing them on top of each other
dfs = [df, new_df1]
result = pd.concat(dfs)
result
````

This method leaves us with many NaN values. You may be wondering, what does the NaN mean? NaN is a common thing you will see with missing/faulty data. We won't talk much about what to do with this, or how properly to take care of it, but know that it is something you must think about when working with a dataset. Here are some examples of how you can fill missing values.

````python
result.fillna(0)
````

````python
result.fillna(np.mean(df['V']))
````

**Another method we will look at is $pd.merge()$**

For those who are familiar with SQL, pandas provides a single function, merge(), for all standard database join operations between DataFrame or named Series objects. There are many types of merges or joins, but today we are going to look at an "inner" join which joins the two dataframes where there is an intersection of keys from both frames.

To get a better idea of the full capabilities of merge you can check here: https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

````python
result = pd.merge(new_df1, new_df2, on="A")
result
````

````python
#We can use the renaming techniques above to rename the columns after merging
result.rename({'B_x' : 'B', 'B_y': 'C'}, axis=1)
````

## Filtering with Pandas <a class="anchor" id="filtering"></a>
By filtering, we can remove data that we are uninterested in. Here are two main ideas we can use filtering for:
- keep data that fall within a range of values to narrow our scope for visualization
- remove faulty data or outliers that skew the perception of our visualizations

As we introduced in our last workshop, the most common format to filter data is as follows:
```
table_name[table_name[column] == value] # Only keep rows with this value in the selected column
table_name[table_name[column] != value] # Only keep rows WITHOUT this value in the selected column
table_name[table_name[column] > value] # Only keep rows with greater values in the selected column
```

For instance, given our dummy dataset we were just looking at, let's say that we want the data in column 'New_Column' that pertains to entries greater than 2.

````python
df = pd.DataFrame({"A": [1, 2, 3, 4],
                   "B": [5, 6, 7, 8], 
                   "C": [1, 2, 3, 4],
                  "alphabet": ['A', 'B', 'C', 'D']})

df
````

````python
df[df['C'] > 2]
````

Now notice if we return df all the rows are still present. If you want access to your filtered table you must save it as a variable or reassign the dataframe to be the filtered table.

````python
df
````

````python
greater_2 = df[df['C'] > 2]
greater_2
````

Here are some examples of things that can also be done with filtering:

````python
#Filter by where column A equals 1
df[df['A'] == 1]
````

````python
#Filter by where column A does not equal 1
df[df['A'] != 1]
````

````python
#Filter by where column B contains even numbers
df[df['B'] % 2 == 0]
````

**Fact: Any conditional statment can be used to filter the dataframe**

You can also filter by multiple conditions by putting each condition in parentheses, and combining them with an ampersand (&). You can separate this into multiple lines, or just put it all together in the same line. For the sake of readibility, we put it in separate lines in this format example:
```
condition1 = table_name[column1] == value1
condition1 = table_name[column2] >= value2
table_name[(condition1) & (condition2)] # Only keep rows with both conditions satisfied
```

````python
condition1 = df['C'] > 3
condition2 = df['alphabet'] == 'D'
df[(condition1) & (condition2)]
````

````python
# both formats work, and generate the same result!
df[(df['C'] > 3) & (df['alphabet'] == 'D')]
````

Now, as a final example, let's say we are trying to find the record with the largest number of entries in column B of the dataframe. We can do this in two steps.

````python
max_entries = np.max(df['B'])
# then, find the row(s) that match `max_entries` in the dataframe
df[df['B'] == max_entries]
````

From this single-record dataframe, we see that the 'alphabet' column where there was the most entries for column 'B' was 'D'.

In sum, using filtering, we can grab certain aspects of data that we want, and see more clearly specific points of interest. Later on, we will see how this can be useful for visualization.

## Grouping with Pandas <a class="anchor" id="grouping"></a>
One thing to note is that the result of calling `.groupby` on a DataFrame is not a Series or DataFrame ‚Äì it is a `pd.core.groupby.generic.DataFrameGroupBy` object. To get this into a usable form, we typically have to use some aggregate function, or use `.apply` or `.filter`.

Here is useful code that will show you what the different "groups" in a .groupby() function will look like without a .apply() or .filter()

````python
grouped_df = df.groupby('Tm')

for key, item in grouped_df:
    print(grouped_df.get_group(key), "\n\n")
````


By grouping, we can stratify our dataset by the unique values of a specific column in our dataset. Then, we can create visualizations based on the different stratas, and compare how they appear similar or different.

When we use the `groupby` function with Pandas, we will get a dataframe where:
- the column you specify to group by will become the index
- there will be 1 row for each unique value in the column
- every other column is aggregated by the function you choose

---

The syntax for groupby is as follows:
```
df.groupby(column).some_function()
```


````python
df.groupby('Tm').max()
df
````

````python
df.groupby('Tm').first()
df
````

````python
df.groupby('Tm').mean()
df
````

What if we wanted the name of the highest scoring player on each team?

````python
df.sort_values('PTS', ascending = False).groupby('Tm').first().head()
````

What if we wanted to find the oldest player on every team?

````python
df.sort_values('Age', ascending = False).groupby('Tm').first().head()
````

### Lambda functions

````python
df.groupby('Tm').agg(lambda x: x.max()).head()
````

````python
df.groupby('Tm').agg(lambda x: x.iloc[0]).head()
````

````python
df.groupby('Tm').agg(lambda x: x.mean()).head()
````
